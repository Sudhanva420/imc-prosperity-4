```python
def calculate_slope(prices_window):
    # prices_window: list of recent mid_prices
    if len(prices_window) < 2:
        return 0
    
    # Simple Linear Regression slope: (N*sum(xy) - sum(x)*sum(y)) / (N*sum(x^2) - (sum(x))^2)
    n = len(prices_window)
    x = np.arange(n)
    y = np.array(prices_window)
    
    slope = (n * np.sum(x*y) - np.sum(x)*np.sum(y)) / (n*np.sum(x**2) - (np.sum(x))**2)
    return slope

```

```python
calculate_slope(ipr0_clean['mid_price_clean'])
```

```python
# Create price bins
ipr0['price_bin'] = pd.qcut(ipr0['mid_price_clean'], q=20)

# Average spread per bin
spread_by_price = ipr0.groupby('price_bin')['spread'].mean()

spread_by_price.plot(figsize=(12,5), title="Avg Spread vs Price Level")
```

```python
# Calculate price change (direction) over the next 100 timestamps
ipr0['future_change'] = ipr0['mid_price'].shift(-100) - ipr0['mid_price']

# Correlation between current spread and future price direction
correlation = ipr0['spread'].corr(ipr0['future_change'])
print(f"Correlation between Spread and Future Price Change: {correlation:.4f}")

# Grouping: Do wide spreads usually precede big moves?
spread_effect = ipr0.groupby('spread')['future_change'].mean()

print("Average price move for each spread width:")
print(spread_effect)
```

```python
import math

STRIKES = {
    "VEV_4000": 4000,
    "VEV_4500": 4500,
    "VEV_5000": 5000,
    "VEV_5100": 5100,
    "VEV_5200": 5200,
    "VEV_5300": 5300,
    "VEV_5400": 5400,
    "VEV_5500": 5500,
    "VEV_6000": 6000,
    "VEV_6500": 6500,
}

def norm_cdf(x):
    return 0.5 * (1 + math.erf(x / math.sqrt(2)))

def bs_call_price(S, K, T, sigma):
    if T <= 0:
        return max(S - K, 0)
    if sigma <= 0:
        return max(S - K, 0)

    d1 = (math.log(S / K) + 0.5 * sigma**2 * T) / (sigma * math.sqrt(T))
    d2 = d1 - sigma * math.sqrt(T)

    return S * norm_cdf(d1) - K * norm_cdf(d2)

def implied_vol_call(market_price, S, K, T):
    intrinsic = max(S - K, 0)

    # invalid / arbitrage-broken price
    if market_price < intrinsic:
        return np.nan

    low, high = 1e-4, 5.0

    for _ in range(60):
        mid = (low + high) / 2
        price = bs_call_price(S, K, T, mid)

        if price > market_price:
            high = mid
        else:
            low = mid

    return (low + high) / 2
```

```python
def build_iv_table(df, tte_days):


    underlying = (
        df[df["product"] == "VELVETFRUIT_EXTRACT"]
        [["timestamp", "mid_price"]]
        .rename(columns={"mid_price": "S"})
    )

    rows = []

    for symbol, K in STRIKES.items():
        opt = (
            df[df["product"] == symbol]
            [["timestamp", "mid_price"]]
            .rename(columns={"mid_price": "option_mid"})
        )

        merged = opt.merge(underlying, on="timestamp", how="inner")

        T = tte_days / 365

        merged["product"] = symbol
        merged["strike"] = K
        merged["tte_days"] = tte_days
        merged["moneyness"] = merged["S"] - K
        merged["relative_moneyness"] = merged["S"] / K - 1

        merged["implied_vol"] = merged.apply(
            lambda row: implied_vol_call(
                row["option_mid"],
                row["S"],
                row["strike"],
                T
            ),
            axis=1
        )

        rows.append(merged)

    return pd.concat(rows, ignore_index=True)
```

```python
plot_df["m_t"] = np.log(plot_df["strike"] / plot_df["S"])
plot_df["v_t"] = plot_df["implied_vol"]

# optional: remove broken low-IV points
plot_df = plot_df[plot_df["v_t"] > 0.05]

# fit parabola globally
x = plot_df["m_t"].values
y = plot_df["v_t"].values

coeffs = np.polyfit(x, y, deg=2)

x_grid = np.linspace(x.min(), x.max(), 300)
y_fit = np.polyval(coeffs, x_grid)

# residuals / detrended IV
plot_df["fitted_iv"] = np.polyval(coeffs, plot_df["m_t"])
plot_df["iv_residual"] = plot_df["v_t"] - plot_df["fitted_iv"]

plt.figure(figsize=(10, 6))

for strike in sorted(plot_df["strike"].unique()):
    temp = plot_df[plot_df["strike"] == strike]
    plt.scatter(
        temp["m_t"],
        temp["v_t"],
        s=8,
        alpha=0.45,
        label=f"strike={strike}"
    )

plt.plot(x_grid, y_fit, linewidth=4, color="black", label="fitted parabola")

plt.xlabel("m_t = log(K / S)")
plt.ylabel("v_t = implied volatility")
plt.title("Volatility Smile: IV vs Moneyness")
plt.legend()
plt.grid(True)
plt.show()
```

```python
STRIKES = {
    "VEV_4000": 4000,
    "VEV_4500": 4500,
    "VEV_5000": 5000,
    "VEV_5100": 5100,
    "VEV_5200": 5200,
    "VEV_5300": 5300,
    "VEV_5400": 5400,
    "VEV_5500": 5500,
    "VEV_6000": 6000,
    "VEV_6500": 6500,
}

def norm_cdf(x):
    return 0.5 * (1 + math.erf(x / math.sqrt(2)))

def bs_call(S, K, T, sigma):
    if T <= 0 or sigma <= 0:
        return max(S - K, 0)
    d1 = (math.log(S / K) + 0.5 * sigma**2 * T) / (sigma * math.sqrt(T))
    d2 = d1 - sigma * math.sqrt(T)
    return S * norm_cdf(d1) - K * norm_cdf(d2)

def implied_vol(price, S, K, T):
    intrinsic = max(S - K, 0)
    if price < intrinsic:
        return np.nan

    low, high = 1e-4, 3
    for _ in range(40):
        mid = (low + high) / 2
        if bs_call(S, K, T, mid) > price:
            high = mid
        else:
            low = mid
    return (low + high) / 2

# underlying
vfe = vfe0
vfe = vfe.rename(columns={"mid_price": "S"})

rows = []

T = 5/365  # round 3

for sym, K in STRIKES.items():
    opt = (
            prices0[prices0["product"] == sym]
            [["timestamp", "mid_price"]]
            .rename(columns={"mid_price": "opt_price"})
        )

    merged = opt.merge(vfe, on="timestamp", how="inner")

    merged["strike"] = K
    merged["product"] = sym
    merged["moneyness"] = merged["S"] - K
    merged["rel_moneyness"] = merged["S"] / K - 1

    merged["iv"] = merged.apply(
        lambda r: implied_vol(r.opt_price, r.S, r.strike, T),
        axis=1
    )

    rows.append(merged)

iv_df = pd.concat(rows, ignore_index=True)

def get_atm_strike(S):
    return min(STRIKES.values(), key=lambda k: abs(S - k))

iv_df["atm_strike"] = iv_df["S"].apply(get_atm_strike)

# get ATM IV per timestamp
atm_df = iv_df[iv_df["strike"] == iv_df["atm_strike"]][["timestamp", "iv"]]
atm_df = atm_df.rename(columns={"iv": "atm_iv"})

iv_df = iv_df.merge(atm_df, on="timestamp", how="left")

iv_df["iv_diff"] = iv_df["iv"] - iv_df["atm_iv"]
```

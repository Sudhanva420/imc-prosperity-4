```python
from statsmodels.tsa.stattools import adfuller
import pandas as pd
import numpy as np

def run_all_checks(df, group_name):
    print(f"\n--- Analysis for Group: {group_name.upper()} ---")
    
    # 1. Pivot the data so each product is a column
    pivot = df.pivot(index='timestamp', columns='product', values='mid_price').fillna(method='ffill')
    products = pivot.columns
    
    # 2. Check for High Correlation (General Relationship)
    corr_matrix = pivot.corr()
    avg_corr = corr_matrix.values[np.triu_indices_from(corr_matrix.values, k=1)].mean()
    print(f"Average Group Correlation: {avg_corr:.4f}")

    # 3. Check for Lead-Lag (Shifted Correlation)
    # Checks if Product A at (t-1) correlates with Product B at (t)
    for i in range(len(products)):
        for j in range(len(products)):
            if i != j:
                lag_corr = pivot[products[i]].shift(1).corr(pivot[products[j]])
                if lag_corr > 0.85:
                    print(f"  [!] LEAD-LAG FOUND: {products[i]} leads {products[j]} (Corr: {lag_corr:.4f})")

    # 4. Check for Stationarity (Mean Reversion)
    # We check if the price itself is mean-reverting (rare) or the spread is
    for i in range(1): # Testing first pair for brevity, can expand
        p1, p2 = products[0], products[1]
        spread = pivot[p1] - pivot[p2]
        result = adfuller(spread.dropna())
        if result[1] < 0.05:
            print(f"  [!] MEAN REVERSION: Spread between {p1} and {p2} is Stationary (p-val: {result[1]:.4f})")

    # 5. Check for Integer Ratios (Arbitrage)
    # Useful for Panels/Pebbles
    for i in range(len(products)):
        for j in range(i+1, len(products)):
            ratios = pivot[products[i]] / pivot[products[j]]
            if ratios.std() < 0.01: # Ratio is almost constant
                print(f"  [!] FIXED RATIO: {products[i]} / {products[j]} = {ratios.mean():.2f}")

# Execute on all DFs
all_dfs = {
    "galaxy": df_galaxy, "sleep": df_sleep, "microchip": df_microchip, 
    "pebbles": df_pebbles, "robots": df_robots, "visors": df_visors, 
    "translators": df_translators, "panels": df_panels, "shakes": df_shakes, "snacks": df_snacks
}

for name, df in all_dfs.items():
    run_all_checks(df, name)
```

```python
import statsmodels.api as sm

def run_deep_stats(df, group_name):
    print(f"\n--- Deep Stats: {group_name.upper()} ---")
    pivot = df.pivot(index='timestamp', columns='product', values='mid_price').fillna(method='ffill')
    products = pivot.columns

    # 1. Self-Correlation (Seasonality/Cycles)
    # Checks if the price repeats itself every N steps
    for prod in products:
        acf = sm.tsa.acf(pivot[prod], nlags=20)
        if np.any(np.abs(acf[1:]) > 0.5):
            lag_val = np.argmax(np.abs(acf[1:])) + 1
            print(f"  [!] CYCLE FOUND: {prod} repeats/trends every {lag_val} ticks.")

    # 2. Basket Stability (The 'Mean' Arbitrage)
    # Does the sum of the 5 products stay flat?
    group_sum = pivot.sum(axis=1)
    if group_sum.std() / group_sum.mean() < 0.005:
        print(f"  [!] BASKET STABILITY: The total price of this group is constant!")

    # 3. Digit/Increment Analysis
    # Checks if prices move in fixed 'steps'
    for prod in products:
        diffs = pivot[prod].diff().dropna()
        unique_diffs = diffs[diffs != 0].abs().unique()
        if len(unique_diffs) < 5:
            print(f"  [!] DISCRETE STEPS: {prod} only moves in increments of {unique_diffs}")

    # 4. Momentum vs Mean Reversion (Hurst Exponent-lite)
    # Is it 'trending' or 'mean-reverting'?
    for prod in products:
        diffs = pivot[prod].diff().dropna()
        # Simple check: do signs flip frequently?
        sign_flips = np.sum(np.diff(np.sign(diffs)) != 0) / len(diffs)
        if sign_flips > 0.65:
            print(f"  [!] STRONG MEAN REVERSION: {prod} (Flipping freq: {sign_flips:.2f})")
        elif sign_flips < 0.35:
            print(f"  [!] STRONG TRENDING: {prod} (Flipping freq: {sign_flips:.2f})")

# Run on the DFs
for name, df in all_dfs.items():
    run_deep_stats(df, name)
```

```python
choco_prices = df_shakes[df_shakes['product'] == 'OXYGEN_SHAKE_CHOCOLATE']['mid_price']
diffs = choco_prices.diff().dropna()

# Check: If the move is BIG, is the reversal more likely?
for min_move in [0, 1, 2, 5]:
    significant_moves = diffs[diffs.abs() >= min_move]
    # Check if next move is opposite sign
    next_moves = diffs.shift(-1).loc[significant_moves.index]
    reversals = (np.sign(significant_moves) != np.sign(next_moves)).sum()
    prob = reversals / len(significant_moves)
    print(f"Min Move {min_move}: Count: {len(significant_moves)} | Rev Prob: {prob:.2%}")
```

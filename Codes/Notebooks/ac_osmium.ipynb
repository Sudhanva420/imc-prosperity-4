```python
# 1. Prices: The market stays where it was (Forward Fill)
price_cols = ['bid_price_1', 'ask_price_1', 'bid_price_2', 'ask_price_2']
aco0[price_cols] = aco0[price_cols].ffill()

# 2. Volumes: If it's missing, it's empty (Fill with Zero)
vol_cols = ['bid_volume_1', 'ask_volume_1', 'bid_volume_2', 'ask_volume_2']
aco0[vol_cols] = aco0[vol_cols].fillna(0)

# 3. Micro-price Logic Fix:
# If L2 is still NaN (e.g. at the very start of the file), 
# make L2 price = L1 price so the weighted average doesn't crash.
aco0['bid_price_2'] = aco0['bid_price_2'].fillna(aco0['bid_price_1'])
aco0['ask_price_2'] = aco0['ask_price_2'].fillna(aco0['ask_price_1'])
```

```python
# Calculate the percentage of time the spread is at the minimum observed value
def get_spread_stats(df):
    
    threshold = 15
    pct_max = (df['spread'] > threshold).mean() * 100
    return pd.Series({
        '% Time at above': f"{pct_max:.2f}%",
        'Min Spread': df['spread'].min(),
        'Avg Spread': df['spread'].mean()
    })

spread_analysis = get_spread_stats(aco0[['spread_l1']].rename(columns={'spread_l1': 'spread'}))

spread_analysis
```

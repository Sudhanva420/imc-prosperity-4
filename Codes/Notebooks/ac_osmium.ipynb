```python
price_cols = ['bid_price_1', 'ask_price_1']

aco0['valid_l1'] = aco0[price_cols].notna().all(axis=1)
```

```python
aco0_valid = aco0[aco0['valid_l1']].copy()
aco0_valid['spread'] = aco0_valid['ask_price_1'] - aco0_valid['bid_price_1']
```

```python
aco0_valid['mid_price'] = (aco0_valid['bid_price_1'] + aco0_valid['ask_price_1']) / 2
```

```python
plt.figure(figsize=(12, 10))

plt.plot(aco0_valid['mid_price'], label='Mid Price')
```

```python
aco0_valid.describe()
```

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(aco0_valid['mid_price'].dropna())
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
```

```python
result = adfuller(aco0_valid['mid_price'].diff().dropna())
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
```

```python
df = aco0_valid.copy()
```

```python
dev = (df['mid_price'] - df['mean_price']).dropna()
```

```python
for w in [20, 50, 100, 200]:
    dev = (df['mid_price'] - df['mid_price'].rolling(w).mean()).dropna()
    print(w, dev.autocorr(lag=1))
```

```python
df['returns'] = df['mid_price'].diff()
df['returns'].autocorr(lag=1)
```

```python
df['spread'].describe()
```

```python
window = 20 
df['mu'] = df['mid_price'].rolling(window=window).mean()
df['std'] = df['mid_price'].rolling(window=window).std()

df['zscore'] = (df['mid_price'] - df['mu']) / df['std']

df['future_return'] = df['mid_price'].shift(-10) - df['mid_price']
print(df[df['zscore'] > 2]['future_return'].mean())
```

```python
print(df[df['zscore'] > 3]['future_return'].mean())
```

```python
print(df[df['zscore'] < -3]['future_return'].mean())
```

```python
plt.figure(figsize=(12, 10))
plt.plot(df['mid_price'], label='Mid Price')
plt.plot(df['mu'], label='Rolling Mean')
plt.legend()
```

```python
for w in [5, 10, 20, 50, 100]:
    df['rolling_mu'] = df['mid_price'].rolling(w).mean()
    df['rolling_std'] = df['mid_price'].rolling(w).std()
    df['z'] = (df['mid_price'] - df['rolling_mu']) / df['rolling_std']
    
    avg_move = df[df['z'] > 2]['future_return'].mean()
    print(f"Window {w}: Avg Move {avg_move:.2f}")
```
